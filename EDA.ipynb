```python
data = pd.read_csv("TimeSeriesDatasetModified.csv")
```

```python
#line graph for bdl 
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')

data.fillna(0, inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['bdl_High'], label='HIGH', color='green')

plt.plot(data['Date'], data['bdl_Low'], label='LOW', color='red')


plt.xlabel('Weekly')
plt.ylabel('BDL Values')
plt.title('Weekly Data from 2018 to 2023')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
