```python
#data preprocessing - box plot to find outliers 
#data cleaning by replacing all outliers with mean values of the column
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df= pd.read_csv("TimeSeriesDataset.csv")
```

```python
#box plot for bdl_high
plt.figure(figsize=(8, 6))  
plt.boxplot(df['bdl_High'], vert=False) 
plt.title('Box Plot of BDL_High')
plt.xlabel('X-axis Label')  
plt.ylabel('Y-axis Label')  
plt.show()  
```

```python

z = np.abs(stats.zscore(df['bdl_High']))
print()

outliers_indices = np.where(z > 2)
mean_df= np.mean(df['bdl_High'])
z[outliers_indices] = mean_df
df.to_csv('bdl_high.csv', index=False)

df['bdl_High'][outliers_indices] = mean_df
plt.figure(figsize=(8, 6))  
plt.boxplot(df['bdl_High'], vert=False) 
plt.title('Box Plot of bdl_High')
plt.xlabel('X-axis Label')  
plt.ylabel('Y-axis Label')  
plt.show() 
```
